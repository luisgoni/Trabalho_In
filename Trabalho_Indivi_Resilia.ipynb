-Temos como desfio uma empresa de desenvolvimento de softwares, onde ela quer um Relatório, através da Analise exploratoria dos dados fornecidos por ela.

-A empresa concluiu que precisava de uma Análise dos dados de um projeto sendo desenvolvido na sua empresa.

##Temos como prioridade gerar um Relatorio de progresso diario, para mostrar o quão produtivo está sendo o trabalho dos funcionarios desta empresa.

-Com as metricas alinhadas podemos responder as perguntas que foram fornceidas para a analise dos dados, que sao elas:

■ Total de Horas Trabalhadas

■ Média Diária de Horas Trabalhadas

■ Total de Bugs Corrigidos

■ Média Diária de Bugs Corrigidos

■ Total de Tarefas Concluídas

■ Média Diária de Tarefas Concluídas

■ Produtividade Diária (Tarefas Concluídas por Hora)



```python
#Começando os trabalhos, importando as bibliotecas, e atribuindo um apelido a elas;
import pandas as pd
import numpy as np
import matplotlib.pyplot as plt
import seaborn as sns
```

- Com as bibliotecas instaladas podemos da inicio a extração de informações atraves dos Dados.

```python
#Criando o Data frame atraves dos dados fornecidos;
dados_Forne = {
    'Dia': ['Segunda-Feira', 'Terça-Feira', 'Quarta-feira', 'Quinta-Feira', 'Sexta-Feira', 'Sábado', 'Domingo'],
    'Horas Trabalhadas': [6, 7, 8, 6, 7, 5, 4],
    'Bugs Corrigidos': [3, 2, 1, 4, 3, 2, 1],
    'Tarefas Concluídas': [5, 4, 6, 4, 5, 3, 2]
}

df_Relatorio = pd.DataFrame(dados_Forne)
df_Relatorio
```

- Checando as Dimensões dos Dados e os espaços em branco no DF.

```python
#Para chegar as dimensões dos dados utilize o .shape ;
Visualizando_Dimen  = df_Relatorio.shape
```

```python
#Antes de commeçar, verificando se a espaços nulos ou ausentes de inforaçoes.
df_Relatorio.info()
#Com a verificação feita podemos partir para a extração dos DAdos.
```

-Com o Data Frame criado podemos Análisar os dados e responder as perguntas que nos foram designadas;


```python
#Perguntas;
# Total de horas trabalhadas ao longo de 1 semana...
p1 = df_Relatorio['Horas Trabalhadas'].sum()

# Média diária de Horas Trabalhads em 1 semana...
p2 = df_Relatorio['Horas Trabalhadas'].mean().round(2)

# Total de Bugs Corrigidos ao longo de 1 semana...
p3 = df_Relatorio['Bugs Corrigidos'].sum()

# Média diária de bugs corrigidos em 1 semana de trabalho...
p4 = df_Relatorio['Bugs Corrigidos'].mean().round(2)

# Total de tarefas concluídas durante 1semana...
p5 = df_Relatorio['Tarefas Concluídas'].sum()

# Média diária de tarefas concluídas...
p6 = df_Relatorio['Tarefas Concluídas'].mean().round(2)

# Produtividade diária (Tarefas concluídas por Hora);
#Criando uma nova coluna com o nome de Produtividade_Diária;
df_Relatorio['Produtividade_Diária'] = df_Relatorio['Tarefas Concluídas'] / df_Relatorio['Horas Trabalhadas']
```

-Respondendo as perguntas e criando as vizualições das  respostas mais facíl.

```python
#Perguntas; 1
# Total de horas trabalhadas ao longo de 1 semana...
p1 = df_Relatorio['Horas Trabalhadas'].sum()
print(f"O total de horas durante 1 semana foi de {p1} Horas semanais de trbalho. ")
#Criando um grafico de barra para melhor observaçao dos dados;
df_Relatorio.plot(y='Horas Trabalhadas',x = 'Dia', kind='bar')
#Melhorndo a visualizaçao.
plt.plot(df_Relatorio['Dia'], df_Relatorio['Horas Trabalhadas'], color='Red', marker='o')
plt.xlabel('Dia')
plt.ylabel('Horas Trabalhadas')
plt.title('Horas Trabalhadas por Dia')
plt.grid(False)
plt.show()

```

- *Com essa informação podemos ver aonde teve mais picos de carga horaria trabalhada e verficar aonde teve menos Horas de trabalho*


-Continuaçao das respostas

```python
#Perguntas 2;
# Média diária de Horas Trabalhads em 1 semana...
p2 = df_Relatorio['Horas Trabalhadas'].mean().round(2)
print(f"A média de horas trabalhadas durante a semana foi de {p2} Horas trabalhadas")
plt.figure(figsize=(6, 4))
sns.barplot(x=['Média'], y=[p2])
plt.title('Média Diária de Horas Trabalhadas')
plt.ylabel('Horas')
plt.show()
```

- Com esse valor obtido , podemos observar que a media diaria de trabalho dessa empresa gira em torno de 6horas 14Minutos .

-Continuação das Resostas;

```python
# Pergunta 3;
# Total de Bugs Corrigidos ao longo de 1 semana...
p3 = df_Relatorio['Bugs Corrigidos'].sum()
print(f'O total de Bugs corrigidos foi de {p3} Bugs por semana. ')
plt.figure(figsize=(10, 6))
plt.plot(df_Relatorio['Dia'], df_Relatorio['Bugs Corrigidos'], color='green', marker='o')
plt.xlabel('Dia')
plt.ylabel('Bugs corrigidos')
plt.title('Bugs corrigidos por dia ')
plt.grid(True)
plt.show()
```

- Com esse dados obtidos conseguimos identificar que tivemos alguns dias em que as correçoes de bugs nao foram bem sucedidas.

-Continuaçao, Visualização das Respostas e dos dados extraidos.

```python
#Pergunta 4;
# Média diária de bugs corrigidos em 1 semana de trabalho...
p4 = df_Relatorio['Bugs Corrigidos'].mean().round(2)
print(f'Com a media de {p4} por semana, podemos observar que tiveram dias que conseguiram corrigir mais bugs que o normal')
print(
f"Com essas informações percebemos que nos dias; Segunda, quinta e Sexta_Feira, foram corrigidos em media de 3 a 4 bugs por dia ")
plt.figure(figsize=(6, 4))
sns.barplot(x=['Média'], y=[p4])
plt.title('Média Diária de Bugs corrigidos ')
plt.ylabel('Dia')
plt.show()
print(df_Relatorio.loc[(df_Relatorio['Bugs Corrigidos'] > p4)])
```

- Com essas informações percebemos que os Bugs corrigidos nao estao diretamente ligados a tarefasconcluidas.

```python
#Pergunta 5 ;
# Total de tarefas concluídas durante 1semana...
p5 = df_Relatorio['Tarefas Concluídas'].sum()
print(f'O total durante a semana foi de {p5} Tarefas concluidas. ')
#Criando um grafico para melhor compreensão do resultado;
plt.figure(figsize=(10, 6))
plt.plot(df_Relatorio['Dia'], df_Relatorio['Tarefas Concluídas'], color='green', marker='o')
plt.xlabel('Dia')
plt.ylabel('Tarefas  corrigidos')
plt.title('Tarefas corrigidas por dia ')
plt.grid(True)
plt.show()
```

- Com o resultado de 29 Tarefas Concluídas por semana, podemos observar que tiveos picos e vales de realizações de tarefas, e que em certos dias, a equipe ja não consegue entregar bons resultados

- Pergunta 6 ;
# Média diária de tarefas concluidas;

```python
#Pergunta 6 ;
# Média diária de tarefas concluídas...
p6 = df_Relatorio['Tarefas Concluídas'].mean().round(2)
print(f'A média diária de tarefas concluidas foram de {p6}')
#Criando os Gráficos;
plt.figure(figsize=(6, 4))
sns.barplot(x=[p6], y=[p6])
plt.title('Média Diária de Tarefas Concluídas')
plt.ylabel('Tarefas Concluídas')
plt.show()
print(f'Podemos Verificar que 3 dias da semana a equipe foi mais eficiente em comparaçao ao longo da semana')
df_Relatorio.loc[df_Relatorio['Tarefas Concluídas'] > p6]
```

-Com essas comparaçoes realizadas , podemos responder a ultima pergunta que foi de Produtividade Diária;
# Pergunta de produtividade diaria;


```python
# Produtividade diária (Tarefas concluídas por Hora);
#Criando uma nova coluna com o nome de Produtividade_Diária;
df_Relatorio['Produtividade_Diária'] = df_Relatorio['Tarefas Concluídas'] / df_Relatorio['Horas Trabalhadas']

#Criando o grafico para melhor visualização desse Dado;
plt.figure(figsize=(10, 5))
sns.barplot(x='Dia', y='Produtividade_Diária', data=df_Relatorio)
plt.title('Produtividade Diária = (Tarefas Concluídas por Hora)')
plt.xlabel('Dia')
plt.ylabel('Tarefas Concluídas por Hora')
plt.show()
```

- Criamos esse grafico de produtividade para que ele possa orientarmos aonde são os dias em que a produtividade por hora foi semelhante ao Tarefas Concluídas
